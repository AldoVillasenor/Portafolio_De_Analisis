```python
import numpy as np
import plotly.express as px
import math
from scipy.stats import chi2,f
```

```python
def dnorm(x,m,s):
    y = (1/s*np.sqrt(2+np.pi))*np.exp(-((x-m)**2)/(2*(s**2)))
    return (y/sum(y))

def dt(x,gl):
    y=(math.gamma((gl+1)/2)/(math.gamma(gl/2)*np.sqrt(gl*np.pi)))*(1/(((x**2)/gl)+1))**(math.gamma((gl+1)/2))
    return (y/sum(y))

def dchi(x,gl):
    return 1/(2*math.gamma(gl/2)*((x/2)**((gl/2)-1))*np.exp(-x/2))
    
```

```python
miu = 0 
sigma = 1
x = np.arange(miu - 4*sigma, miu + 4*sigma, 0.01)
y = dnorm(x,miu, sigma)
px.line(x = x, y = y, width=1200, height=500)
```

```python
gl = 5  # Grados de libertad
sigma = np.sqrt(gl/(gl-2))
x = np.arange( -4*sigma, 4*sigma, 0.01)
y = dt(x,gl)
fig1 = px.line(x = x , y = y, width=1200, height=500)
fig1.show()
```

```python
gl = 10  # Grados de libertad
sigma = np.sqrt(2 * gl)
x = np.arange( -1*sigma, 6*sigma, 0.01)
y = chi2.pdf(x,gl)
fig2 = px.line(x = x , y = y, width=1200, height=500)
fig2.show()
```

```python
v1 = 6
v2 = 10
sigma = np.sqrt(2)*v2*np.sqrt(v2+v1-2)/(np.sqrt(v2-4)*(v2-2)*np.sqrt(v1))
x = np.arange( 0, miu + 5*sigma, 0.01)
y = f.pdf(x,v1, v2)
fig3 = px.line(x = x , y = y, width=1200, height=500)
fig3.show()
```
